Question (Based on filtering + Value Counts)

1. For Speeding, Were Men or Women stopped more often?

```python
data.head()
```

```python
data[data.violation == "Speeding"].driver_gender.value_counts()
```

```python
import matplotlib.pyplot as plt
```

```python
# Filtering the dataset for "Speeding" violations and counting gender occurrences
gender_counts = data[data['violation'] == "Speeding"]['driver_gender'].value_counts()
```

```python
# Plotting the data
plt.figure(figsize=(8, 5))
gender_counts.plot(kind='bar', color=['skyblue', 'pink'])

# Customizing the plot
plt.title("Number of Drivers Stopped for Speeding by Gender")
plt.xlabel("Driver Gender")
plt.ylabel("Number of Drivers")
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Showing the plot
plt.tight_layout()
plt.show()
```
